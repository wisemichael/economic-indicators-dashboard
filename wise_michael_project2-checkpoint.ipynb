# File Loading and Preprocessing

## Data Preprocessing & Cleaning

###  Dataset Overview
The dataset consists of multiple CSV files, each representing a different **economic indicator** 
(such as inflation, unemployment, and wages). Some indicators had different date ranges and formats, 
which required preprossing into into a single, more structured dataset.

### Data Cleaning:
**Standardizing the Timeframe:**
   - Converted the `observation_date` column to a **datetime format**.
   - Filtered the dataset to include **only records from 1950 onward**.

**Renaming Columns:**
   - Column names were standardized based on filenames.
   - Spaces were replaced with underscores to ensure consistency.

**Merging Datasets:**
   - All files were merged on `observation_date` using an **outer join** to keep all available data.

**Handling Missing Values:**
   - Identified missing values using a **heatmap visualization**.
   - Ensured `observation_date` remained consistent after merging.

**Saving Cleaned Data:**
   - The final cleaned dataset was saved as **economic_data_cleaned.csv** for further analysis.

This cleaned dataset was then used to create **interactive economic visualizations** in the dashboard.

```python
# Directory.
data_dir = r"C:\Users\wisem\Desktop\Data Visualization\Week 10"

# List of dataset files.
files = [
    "average_hourly_earnings.csv",
    "Consumer Price Index FRED.csv",
    "CPI_monthly_seasonally.csv",
    "disposable_personal_income_b.csv",
    "employment_cost_index_wage_growth.csv",
    "federal_funds_interest_rates.csv",
    "median_personal_income_US.csv",
    "PCE_durable_goods.csv",
    "PCE_energy_goods_services.csv",
    "PCE_non_durable_goods.csv",
    "PCE_services.csv",
    "personal_savings_rate.csv",
    "recession_indicator.csv",
    "unemployment_rate.csv"
]

# Creates an empty dictionary to store DataFrames.
dataframes = {}

# Converts `observation_date` to datetime, and filter from 1950 onward.
cutoff_date = pd.to_datetime("1950-01-01")

for file in files:
    file_path = os.path.join(data_dir, file)

    try:
        df = pd.read_csv(file_path)
        
        # Converts 'observation_date' to datetime.
        if 'observation_date' in df.columns:
            df['observation_date'] = pd.to_datetime(df['observation_date'], errors='coerce')
            df = df[df['observation_date'] >= cutoff_date]  # Keeps only dates from 1950 onward.

        # Renames second column based on filename.
        col_name = file.replace(".csv", "").replace(" ", "_").lower()
        df = df.rename(columns={df.columns[1]: col_name})

        dataframes[file] = df
        print(f" Loaded and cleaned: {file}")

    except Exception as e:
        print(f" Error reading {file}: {e}")

# Merges alls datasets on `observation_date`.
economic_data = None
for df in dataframes.values():
    if economic_data is None:
        economic_data = df
    else:
        economic_data = pd.merge(economic_data, df, on='observation_date', how='outer')

# Ensures 'observation_date' is datetime after merge.
economic_data['observation_date'] = pd.to_datetime(economic_data['observation_date'], errors='coerce')

# Saves cleaned dataset.
economic_data.to_csv("economic_data_cleaned.csv", index=False)
print("\n Final cleaned dataset saved as 'economic_data_cleaned.csv'")

# Reloads and verify `observation_date` is present and still datetime.
df = pd.read_csv("economic_data_cleaned.csv", parse_dates=['observation_date'])

# Prints final data types and missing values.
print("\n Final Data Types:")
print(df.dtypes)

print("\n Date Range:", df['observation_date'].min(), "to", df['observation_date'].max())

print("\n Missing Values:")
print(df.isnull().sum())

print("\n Columns after reloading:", df.columns)

# Plots missing values heatmap.
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cmap="viridis", cbar=False, yticklabels=False)
plt.title("Missing Data Heatmap")
plt.show()
```

```python
# Reloads cleaned dataset.
df = pd.read_csv("economic_data_cleaned.csv", parse_dates=['observation_date'])

# Ensures 'observation_date' is retained.
if 'observation_date' not in df.columns:
    df.reset_index(inplace=True)

# Handles missing values: forward-fill, then back-fill as a last resort.
df.ffill(inplace=True)
df.bfill(inplace=True)

# Sets observation_date as index for analysis.
df.set_index('observation_date', inplace=True)

# Drops columns where more than 50% of data is missing.
threshold = 0.5 * len(df)
df.dropna(thresh=threshold, axis=1, inplace=True)

# Saves final dataset.
df.to_csv("economic_data_final.csv")
print(" Final cleaned dataset saved as 'economic_data_final.csv'")

# Reloads and verify data information.
df = pd.read_csv("economic_data_final.csv", parse_dates=['observation_date'])
print(" Data reloaded successfully")
print(" Final Column List:", df.columns.tolist())
print(" Final Data Types:\n", df.dtypes)
print(" Missing Values After Processing:\n", df.isnull().sum())

display(df.head())
```

# Visualizations

## Visualizing Economic Indicators

The following visualizations help analyze trends in key economic metrics, including:
- Average Hourly Earnings
- Consumer Price Index (Inflation)
- Disposable Personal Income
- Federal Funds Interest Rates
- Unemployment Rate

```python
# Subset of indicators as selected column list.
selected_columns = ['average_hourly_earnings', 'consumer_price_index_fred', 
                    'unemployment_rate', 'federal_funds_interest_rates']

plt.figure(figsize=(12, 6))

# Normalizes all numerical columns between 0 and 1.
df_normalized = df.copy()
for col in df.columns:
    if col != "observation_date":  # Skip the date column
        df_normalized[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

# Plots omly selected indicators.
for col in selected_columns:
    plt.plot(df['observation_date'], df_normalized[col], label=col, linewidth=1.5)

# Adjusts legend to remain outside of visualization.
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10, frameon=False)

plt.title("Normalized Economic Indicators Over Time")
plt.xlabel("Year")
plt.ylabel("Normalized Value (0-1)")
plt.grid(True)

# Adds shaded blocks for periods of recession.
if 'recession_indicator' in df.columns:
    recession_df = df[df['recession_indicator'] == 1]  # Get recession periods
    start = None  # Track recession start

    for i in range(len(recession_df)):
        if start is None:  
            start = recession_df.iloc[i]['observation_date']  # Set start date
        # If the next date is not consecutive, close block
        if i == len(recession_df) - 1 or (recession_df.iloc[i+1]['observation_date'] - recession_df.iloc[i]['observation_date']).days > 30:
            end = recession_df.iloc[i]['observation_date']
            plt.axvspan(start, end, color='gray', alpha=0.3)  # Shaded block
            start = None  # Reset

plt.tight_layout()
plt.show()
```

```python
# Defines valid columns (exclude those with all NaN values).
valid_columns = [col for col in selected_columns if df[col].notna().sum() > 0]

# Sets up subplots as 3 rows 2 columns.
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 12))
fig.suptitle("Distribution of Economic Indicators", fontsize=14)

# Flattens axes for easy iteration.
axes = axes.flatten()

# Loops through selected indicators and plots histograms.
for i, col in enumerate(valid_columns[:len(axes)]):
    axes[i].hist(df[col].dropna(), bins=30, edgecolor='black', alpha=0.8, log=False)
    axes[i].set_title(col.replace("_", " ").title(), fontsize=12)
    axes[i].set_xlabel("Value")  # Actual values
    axes[i].set_ylabel("Frequency")  # Frequency of occurrences
    axes[i].grid(True, linestyle='--', alpha=0.6)  # Light grid for readability

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust spacing to prevent overlap.
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Interpolates missing values using linear method.
df[selected_columns] = df[selected_columns].interpolate(method='linear', limit_direction='forward', axis=0)

# Defines a correlation matrix using selected columns.
corr = df[selected_columns].corr()

# Only displays lower triangle for simplisticy.
mask = np.triu(np.ones_like(corr, dtype=bool))

df_filled = df[selected_columns].fillna(method='ffill')  # Forward-fill missing values
corr = df_filled.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, mask=mask, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Lower Triangle Correlation Heatmap of Economic Indicators")
plt.show()
```

Above, we utilized multiple data visualizations to provide a comprehensive understanding of economic trends over time. The selected visualizations help identify patterns, correlations, and distributions of key economic indicators, allowing for deeper insight into various economic components.

This ensures that indicators with different scales (e.g., unemployment rate vs. inflation) can be compared effectively. The overlay of recession periods (shaded gray) further highlights how economic downturns impact these indicators, showing spikes in unemployment and interest rates during recessions, while consumer price index trends downward in periods of economic recovery.

The histogram plots illustrate the frequency distribution of several economic indicators, such as average hourly earnings, unemployment rate, interest rates, and CPI. These distributions provide insight into the variability and common ranges of values for each metric. For instance, the concentration of unemployment rates around certain values helps identify stable economic periods versus times of extreme fluctuation.

Lower Triangle Correlation Heatmap:
The correlations matrix highlights the relationships between different economic indicators and the strenghs between them. Negative correlations, such as the strong inverse relationship between consumer price index and average hourly earnings (-0.82), indicate that as prices increase, real wages struggle to keep pace. Similarly, the weak correlation between unemployment and interest rates suggests a more complex interaction influenced by external economic policies. This heatmap helps us quickly interpret which indicators move together and which may have opposing trends.

These visualizations were chosen in order to enhance the explanatory power of our data, enabling end users to explore the structure and relationships of various economic indicators.

# Dash Application

```python
# Loads the cleaned dataset.
df = pd.read_csv("economic_data_cleaned.csv", parse_dates=["observation_date"])

# Historical recession periods for toggle feature.
recession_periods = [
    ("1973-11-01", "1975-03-01"),
    ("1980-01-01", "1980-07-01"),
    ("1981-07-01", "1982-11-01"),
    ("1990-07-01", "1991-03-01"),
    ("2001-03-01", "2001-11-01"),
    ("2007-12-01", "2009-06-01"),
    ("2020-02-01", "2020-04-01")
]

# Available economic indicators.
available_indicators = [col for col in df.columns if col != "observation_date"]

# Creates Dash app.
app = dash.Dash(__name__)

# Layout of the Dash app and dcc. tags.
app.layout = html.Div([
    html.H1("Economic Indicators Dashboard", style={'textAlign': 'center'}),

    # Dropdown for selecting economic indicators
    dcc.Dropdown(
        id="indicator-dropdown",
        options=[{"label": col.replace("_", " ").title(), "value": col} for col in available_indicators],
        value=["pce_services"],  # Default column displayed
        multi=True
    ),

    # Selection buttons for chart type
    dcc.RadioItems(
        id="chart-type",
        options=[
            {"label": " Line Chart", "value": "line"},
            {"label": " Bar Chart", "value": "bar"}
        ],
        value="line",
        inline=True
    ),

    # Annual range slider bar.
    dcc.RangeSlider(
        id="year-slider",
        min=df["observation_date"].dt.year.min(),
        max=df["observation_date"].dt.year.max(),
        step=1,
        value=[df["observation_date"].dt.year.min(), df["observation_date"].dt.year.max()],
        marks={i: str(i) for i in range(df["observation_date"].dt.year.min(), df["observation_date"].dt.year.max(), 10)}
    ),

    # Toggle for recession markers.
    dcc.Checklist(
        id="recession-toggle",
        options=[{"label": " Show Recession Periods", "value": "show"}],
        value=[],  # Default is OFF
        inline=True
    ),

    # Time-series graph initialization.
    dcc.Graph(id="time-series-graph"),
])

# Callback that updates graph with dual y-axis for various column values (percents vs. raw dollars).
@app.callback(
    Output("time-series-graph", "figure"),
    [Input("indicator-dropdown", "value"),
     Input("chart-type", "value"),
     Input("year-slider", "value"),
     Input("recession-toggle", "value")]
)
def update_graph(selected_indicators, chart_type, year_range, show_recession):
    if not selected_indicators:
        return dash.no_update

    # Filters data by selected year range.
    df_filtered = df[(df["observation_date"].dt.year >= year_range[0]) & 
                     (df["observation_date"].dt.year <= year_range[1])].copy()

    # Fills missing values with interpolation (due to missing data with certain columns).
    df_filtered = df_filtered.set_index("observation_date")
    df_filtered = df_filtered.ffill().interpolate(method='linear').bfill().reset_index()

    # Creates figure.
    fig = go.Figure()

    for indicator in selected_indicators:
        if indicator not in df_filtered.columns:
            continue  # Skip if indicator is not in dataset

        if df_filtered[indicator].isnull().all():
            continue  # Skip indicators that are completely missing

        if chart_type == "line":
            trace = go.Scatter(
                x=df_filtered["observation_date"], 
                y=df_filtered[indicator], 
                mode="lines", 
                name=indicator.replace("_", " ").title(),
                connectgaps=True  # Ensures missing values are connected
            )
        elif chart_type == "bar":
            trace = go.Bar(
                x=df_filtered["observation_date"], 
                y=df_filtered[indicator], 
                name=indicator.replace("_", " ").title()
            )

        fig.add_trace(trace)

    # Adjusts x-axis to allow for spacial clarity using 45 degree ticks.
    fig.update_xaxes(tickangle=-45, tickmode="auto")

    # Layout of figure.
    fig.update_layout(
        title="Economic Trends Over Time",
        xaxis_title="Year",
        yaxis_title="Economic Indicator Value",
        legend_title="Indicators",
        barmode="group" if chart_type == "bar" else None  # Ensures grouped bars
    )

    # Adds recession markers toggle.
    if "show" in show_recession:
        for start, end in recession_periods:
            fig.add_vrect(
                x0=start, x1=end,
                fillcolor="gray", opacity=0.3,
                layer="below", line_width=0
            )

    return fig

# Runs the dash app.
if __name__ == "__main__":
    app.run_server(debug=True)
```

The analysis of economic indicators over time reveals several notable trends and relationships. Inflation (CPI - Monthly Seasonally Adjusted) and unemployment exhibit a cyclical relationship, reflecting fluctuations in the economic hisotry of the US . Periods of economic expansion are often accompanied by steady wage growth and declining unemployment, whereas downturns trigger rising inflation and job losses. Disposable Personal Income has shown a consistent upward trajectory, suggesting an overall increase in earnings and consumer expenditures during these periods. However, interest rates tend to be highly volatile, with sharp increases occurring in response to economic uncertainty, particularly during financial recessions.

Recession periods provide critical insight into the dynamic between key indicators. During downturns, disposable income often declines, while unemployment rates surge, highlighting the immediate impact of economic slowdowns on individuals. The Federal Reserve’s interest rate adjustments play a significant role in mitigating these effects, as rates are frequently lowered to encourage borrowing and stimulate growth, followed by increases once the economy reaches a period of "normalcy." Savings rates also tend to shift in response to economic uncertainty, displaying consumer diligence and frugality during difficult times and a more laize-affair approach when the economy is more robust.

The dashboard is a useful tool for this type of analysis, by allowing users to compare both monetary values and percentage-based indicators simultaneously. The dual-axis outlay ensures that high-value economic indicators (e.g., wages, disposable income, and consumer spending) remain comparable with percent-based metrics (e.g., unemployment, interest rates, and savings rates), preventing key insights from being overshadowed. The ability to toggle recession periods also provides a deeper view of the potential associations of economic indicators and resulting policy change. By combining these indicators into a formalized Dash application, this is a powerful tool for that could be used for identifying economic trends and educating consumers. Through this, we can draw meaningful conclusions about financial stability, market behavior, and policy effectiveness throughout U.S. economic history.
